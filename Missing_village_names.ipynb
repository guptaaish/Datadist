```python
import pandas as pd
```

```python
df_1 = pd.read_excel('Tab_Villages_Mapping, Krishna District, 25 July 2015.xlsx')
df_2 = pd.read_excel('Tab_Villages_Mapping, Krishna District, 25 July 2015.xlsx',sheet_name=1)
```

```python
df_1.head(10),df_2.head(10)
```

```python
df_1.info()
```

```python
df_2.info()
```

```python
df_2['Survey Start Date'] = pd.to_datetime(df_2['Survey Start Date'])
df_2['Survey End Date'] = pd.to_datetime(df_2['Survey End Date'])
```

```python
df_2.head(10)
```

```python
df_2.info()
```

```python
df_3 = pd.merge(df_1,df_2,on= 'Tab No',how= 'left')
df_3.info()
```

```python
df_3 = df_3.drop(['AC Name_x', 'Mandal Name_x', 'Village Name_x'],axis=1)
```

```python
df_3.info()
```

```python
df_4 = df_3[(df_3['Survey Date']>=df_3['Survey Start Date']) & (df_3['Survey Date']<=df_3['Survey End Date'])]
```

```python
df_5 = df_4.drop(['Survey Start Date', 'Survey End Date'],axis = 1)
df_5.columns = ['Response No','Tab No','Survey Date','AC Name','Mandal Name','Village Name']
```

```python
df_5.info()
```

```python
df_7 = df_5.drop_duplicates()
```

```python
df_7.info()
```

```python
df_6 = pd.merge(df_1,df_7,on= ['Response No','Tab No','Survey Date'],how= 'left')
df_6 = df_6.drop(['AC Name_x', 'Mandal Name_x', 'Village Name_x'],axis=1)
df_6.columns = (df_7.columns)
df_6.info()
```
